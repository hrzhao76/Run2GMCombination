```python
def genTemplate(syst:str):
    systSubCategory = syst.split('_')[0]
    systBlock = \
    f'Systematic: "{syst}"\n' + \
    f'  Title: "{syst}"\n' + \
    f'  Type: HISTO \n' + \
    f'  Smoothing: 40 \n' + \
    f'  HistoNameSufUp: "_{syst}__1up"\n' + \
    f'  HistoNameSufDown: "_{syst}__1down"\n' + \
    f'  Symmetrisation: TWOSIDED \n' + \
    f'  Category: Experimental \n' + \
    f'  SubCategory: "{systSubCategory}" \n' + \
    '\n'
    
    return systBlock
```

```python
with open("experi_systBlock.txt", 'w') as file:

    for syst in experimentSyst:
        split_eql = syst.split(" = ")
        split_colon = split_eql[1].split(":")
        syst = split_colon[0]
        file.write(genTemplate(syst))
```
